```python
import pandas as pd
import numpy as np
import sys
import os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..', 'src')))
from time_series_quality import TimeSeriesQuality
```

```python
# Carregar dados de exemplo
data = pd.DataFrame({
    "date": pd.date_range(start="2024-01-01", periods=10, freq="D"),
    "value": [1, 2, 3, np.nan, 5, 6, np.nan, 8, 9, 100]
})

# Criando a instância da classe TimeSeriesQuality
ts_quality = TimeSeriesQuality(data, "date", "value")

# Verificando valores faltantes
print(ts_quality.check_missing_values())

# Preenchendo valores faltantes
ts_quality.fill_missing_values(method="median")

# Identificando anomalias com Z-score
anomalies = ts_quality.detect_anomalies(method="zscore", threshold=2)
print("Anomalias (Z-score):\n", anomalies)

# Plotando os dados
ts_quality.plot_data()
```
